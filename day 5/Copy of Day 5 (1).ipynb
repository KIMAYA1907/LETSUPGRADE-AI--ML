```
 def replace_char(y):
    invalid_list = ["+","&","$","#","@"]
    for i in invalid_list:
        y = y.replace(i," ")
        y = y.lower()
    print("remove all invalid syntax")
    print("result is:",y)

string1 = "Dr Drashan ingale@#trainer@AIML"
print("string is:",string1)
replace_char(string1)
```
