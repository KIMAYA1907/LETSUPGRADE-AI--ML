Create 8×8 matrix and fill it with checker board pattern

```
 import numpy as np
x = np.ones((3,3))
print("Checkerboard pattern:")
x = np.zeros((8,8),dtype=int)
x[1::2,::2] = 1
x[::2,1::2] = 1
print(x)
```

Normalise 5×5 random matrix

```
 import numpy as np
x= np.random.random((5,5))
xmax, xmin = x.max(), x.min()
x = (x - xmin)/(xmax - xmin)
print("After normalization:")
print(x)
```

How to find comman value between two arrays

```
 import numpy as np
array1 = np.array([0, 10, 20, 40, 60])
print("Array1: ",array1)
array2 = [10, 30, 40]
print("Array2: ",array2)
print("Common values between two arrays:")
print(np.intersect1d(array1, array2))
```

How to get dates of yesterday today and tomorrow

```
 import datetime 
today = datetime.date.today()
yesterday = today - datetime.timedelta(days = 1)
tomorrow = today + datetime.timedelta(days = 1) 
print('Yesterday : ',yesterday)
print('Today : ',today)
print('Tomorrow : ',tomorrow)
```

Consider two random arrays and check if they are equal

```
 
an_array = np.array([[1,2],[3,4]])
another_array = np.array([[1,2],[3,4]])
 
comparison = an_array == another_array
equal_arrays = comparison.all()
 
print(equal_arrays)
```

Create random vector of size 10 and replace the maximum value by 0

```
 import numpy as np
x = np.random.random(10)
 
print(x)
x[x.argmax()] = 0
print("Maximum value replaced by 0:")
print(x)
```

Substract the mean of each row of matrix

```
 
X = np.random.rand(5, 5)
 
Y = X - X.mean(axis=1, keepdims=True)
print(Y)
```

 Consider a vector,how to add 1 to each element indexed by a 2nd vector

```
 Z = np.ones(10)
I = np.random.randint(0,len(Z),20)
Z += np.bincount(I, minlength=len(Z))
print(Z)
```

```
 x = np.arange(7)
np.random.shuffle(x)
n = 1
print (x[np.argsort(x)[-n:]])
```

How to create a record array from regular array

```
 a1=np.array([1,2,3,4])
a2=np.array(['knp','belll','green','numpy'])
a3=np.array([12.20,15,20,40])
result= np.core.records.fromarrays([a1, a2, a3],names='a,b,c')
print(result[0])
print(result[1])
print(result[2])
```
